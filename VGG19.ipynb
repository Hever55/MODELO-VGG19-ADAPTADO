```python
import tensorflow as tf
from tensorflow.keras import layers, models
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.preprocessing.image import ImageDataGenerator
import matplotlib.pyplot as plt
from tensorflow.keras import regularizers
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau
from google.colab import drive
import numpy as np
from sklearn.metrics import confusion_matrix
import seaborn as sns
import pickle # Import the pickle library

# Montar Google Drive
drive.mount('/content/drive')

def build_improved_vgg19_transfer(input_shape=(256, 256, 3), num_classes=3):
    """
    Modelo VGG19 con transfer learning mejorado
    """
    # Cargar VGG19 sin las capas de clasificación
    base_model = tf.keras.applications.VGG19(
        weights='imagenet',
        include_top=False,
        input_shape=input_shape
    )

    # Fine-tuning: Descongelar las últimas capas convolucionales
    # Congelar los primeros 15 bloques/capas
    for layer in base_model.layers[:15]:
        layer.trainable = False
    # Hacer entrenables las últimas capas
    for layer in base_model.layers[15:]:
        layer.trainable = True

    # Construir el modelo completo
    inputs = tf.keras.Input(shape=input_shape)

    # Preprocesamiento necesario para VGG19 (muy importante)
    x = tf.keras.applications.vgg19.preprocess_input(inputs)

    # Usar el modelo base
    x = base_model(x)

    # Agregar clasificador mejorado
    x = layers.GlobalAveragePooling2D()(x)  # Mejor que Flatten para transfer learning

    # Clasificador más potente
    x = layers.Dense(512, activation='relu')(x)
    x = layers.BatchNormalization()(x)  # Añadir BatchNormalization
    x = layers.Dropout(0.5)(x)  # Aumentar dropout

    x = layers.Dense(256, activation='relu')(x)
    x = layers.BatchNormalization()(x)
    x = layers.Dropout(0.3)(x)

    outputs = layers.Dense(num_classes, activation='softmax')(x)

    model = tf.keras.Model(inputs, outputs)
    return model

# Rutas de datos
ruta_datos_entrenamiento = '/content/drive/MyDrive/entrenamiento'
ruta_datos_validacion = '/content/drive/MyDrive/validacion'
ruta_datos_prueba = '/content/drive/MyDrive/prueba'

# Tamaño de la imagen y batch size
img_size = (256, 256)
batch_size = 32  # Reducir batch size para mejor generalización

# Preprocesamiento mejorado: permitir rescale dentro del modelo para usar preprocess_input de VGG19
train_datagen = ImageDataGenerator(
    # No usar rescale aquí porque usamos preprocess_input de VGG19
    rotation_range=30,  # Aumentar rotación
    width_shift_range=0.2,
    height_shift_range=0.2,
    shear_range=0.2,
    zoom_range=0.3,  # Aumentar zoom
    horizontal_flip=True,
    vertical_flip=True,  # Añadir flip vertical
    fill_mode='nearest',
    brightness_range=[0.8, 1.2]  # Añadir variación de brillo
)

val_datagen = ImageDataGenerator()  # Sin rescale
test_datagen = ImageDataGenerator()  # Sin rescale

# Cargar datos
train_generator = train_datagen.flow_from_directory(
    ruta_datos_entrenamiento,
    target_size=img_size,
    batch_size=batch_size,
    class_mode='categorical'
)

val_generator = val_datagen.flow_from_directory(
    ruta_datos_validacion,
    target_size=img_size,
    batch_size=batch_size,
    class_mode='categorical'
)

test_generator = test_datagen.flow_from_directory(
    ruta_datos_prueba,
    target_size=img_size,
    batch_size=batch_size,
    class_mode='categorical'
)

# Construir el modelo mejorado
model = build_improved_vgg19_transfer(input_shape=(256, 256, 3), num_classes=3)

# Compilar con optimizador mejorado
model.compile(
    optimizer=Adam(learning_rate=0.0001),  # Reducir learning rate
    loss='categorical_crossentropy',
    metrics=['accuracy']
)

# Resumen del modelo
model.summary()

# Número de pasos por época
train_steps = train_generator.samples // batch_size
val_steps = val_generator.samples // batch_size
test_steps = test_generator.samples // batch_size

# Callbacks mejorados
early_stopping = EarlyStopping(
    monitor='val_accuracy',  # Monitorear accuracy en vez de loss
    patience=15,
    restore_best_weights=True,
    verbose=1
)

reduce_lr = ReduceLROnPlateau(
    monitor='val_accuracy',  # Monitorear accuracy en vez de loss
    factor=0.5,  # Reducción más suave
    patience=5,
    min_lr=0.00001,
    verbose=1
)

# Entrenar el modelo
history = model.fit(
    train_generator,
    steps_per_epoch=train_steps,
    validation_data=val_generator,
    validation_steps=val_steps,
    epochs=50,  # Aumentar número de épocas
    callbacks=[early_stopping, reduce_lr],
    verbose=1
)

# Evaluar el modelo
loss_val, accuracy_val = model.evaluate(val_generator)
print(f"Pérdida de validación: {loss_val:.4f}")
print(f"Precisión de validación: {accuracy_val:.4f}")

loss_test, accuracy_test = model.evaluate(test_generator)
print(f"Pérdida de prueba: {loss_test:.4f}")
print(f"Precisión de prueba: {accuracy_test:.4f}")

# Redefinir el generador de prueba SIN mezclar
test_datagen = ImageDataGenerator()
test_generator = test_datagen.flow_from_directory(
    ruta_datos_prueba,
    target_size=img_size,
    batch_size=batch_size,
    class_mode='categorical',
    shuffle=False
)

# Obtener predicciones
test_generator.reset()
predictions = model.predict(test_generator, steps=len(test_generator), verbose=1)
predicted_classes = np.argmax(predictions, axis=1)

# Obtener etiquetas verdaderas
true_classes = test_generator.classes
class_labels = list(test_generator.class_indices.keys())

# Calcular métricas adicionales
from sklearn.metrics import cohen_kappa_score, matthews_corrcoef, balanced_accuracy_score

kappa = cohen_kappa_score(true_classes, predicted_classes)
mcc = matthews_corrcoef(true_classes, predicted_classes)
balanced_acc = balanced_accuracy_score(true_classes, predicted_classes)

print("\nMétricas adicionales:")
print(f"Cohen's Kappa: {kappa:.4f}")
print(f"Matthews Correlation Coefficient: {mcc:.4f}")
print(f"Balanced Accuracy: {balanced_acc:.4f}")

# Curvas ROC y AUC
from sklearn.preprocessing import label_binarize
from sklearn.metrics import roc_curve, auc

# Binarizar las etiquetas para multiclase
n_classes = len(class_labels)
y_test_bin = label_binarize(true_classes, classes=np.arange(n_classes))

# Calcular curvas ROC y áreas para cada clase
fpr = dict()
tpr = dict()
roc_auc = dict()

plt.figure(figsize=(10, 8))
colors = ['blue', 'green', 'red']  # Ajustar según número de clases

for i in range(n_classes):
    fpr[i], tpr[i], _ = roc_curve(y_test_bin[:, i], predictions[:, i])
    roc_auc[i] = auc(fpr[i], tpr[i])
    plt.plot(fpr[i], tpr[i], color=colors[i], lw=2,
             label=f'Clase {class_labels[i]} (AUC = {roc_auc[i]:.2f})')

plt.plot([0, 1], [0, 1], 'k--', lw=2)
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('Tasa de Falsos Positivos')
plt.ylabel('Tasa de Verdaderos Positivos')
plt.title('Curvas ROC por Clase')
plt.legend(loc="lower right")
plt.show()

# AUC Promedio
macro_roc_auc = np.mean(list(roc_auc.values()))
print(f"\nAUC Promedio: {macro_roc_auc:.4f}")

# Calcular matriz de confusión
cm = confusion_matrix(true_classes, predicted_classes)

# Visualización profesional
plt.figure(figsize=(12, 10))
sns.heatmap(cm,
            annot=True,
            fmt='d',
            cmap='Blues',
            cbar=False,
            xticklabels=class_labels,
            yticklabels=class_labels,
            annot_kws={'size': 15})

plt.xticks(rotation=45, fontsize=12)
plt.yticks(rotation=0, fontsize=12)
plt.xlabel('Predicciones', fontsize=14, labelpad=15)
plt.ylabel('Etiquetas Reales', fontsize=14, labelpad=15)
plt.title('Matriz de Confusión - Modelo VGG19 Mejorado', fontsize=16, pad=20)

# Añadir precisión total
total_accuracy = np.trace(cm) / np.sum(cm)
plt.text(0.5, -0.15,
         f'Precisión Total: {total_accuracy:.2%}',
         ha='center',
         va='center',
         transform=plt.gca().transAxes,
         fontsize=12)

plt.tight_layout()
plt.show()

# Métricas adicionales (opcional)
from sklearn.metrics import classification_report
print("\nReporte de Clasificación:")
print(classification_report(true_classes, predicted_classes, target_names=class_labels))

# Gráfico de métricas de entrenamiento
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
plt.plot(history.history['accuracy'], label='Entrenamiento')
plt.plot(history.history['val_accuracy'], label='Validación')
plt.title('ACC durante Entrenamiento')
plt.ylabel('ACC')
plt.xlabel('Época')
plt.legend()

plt.subplot(1, 2, 2)
plt.plot(history.history['loss'], label='Entrenamiento')
plt.plot(history.history['val_loss'], label='Validación')
plt.title('Pérdida durante Entrenamiento')
plt.ylabel('Pérdida')
plt.xlabel('Época')
plt.legend()
plt.show()
```
